```python
import pandas as pd
#pandas is a library function and it related to the database also named as pd short form
```

```python
dataset=pd.read_csv("Salary_Data.csv")#In pd.read_csv function under pandas salary_data.csv is called and assigned to the dataset variable
```

```python
dataset #displaying the dataset
```

```python
independent=dataset[["YearsExperience"]]#Assigning dataset of specific yearsexperience as a input in independant variable(splitting i/p)
independent#displaying independent variable
```

```python
dependant=dataset[["Salary"]]#Assigning dataset of specific salary as a output in dependant variable(splitting output)
dependant #displaying dependent variable
```

```python
from sklearn.model_selection import train_test_split#creating a model using a sklearn library function and model selection as a folder
importing train and test set as split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependant,test_size=0.30,random_state=0)
#after importing the value should be input training set and input test set and output training set and test set 
#like X_train,X_test,Y_train,Y_test= assigned to train_test_split as(train is independant and test as dependant also test value size=0.30
#and random_state=0 this all are assigned to X_train,X_test,Y_train,Y_test
#in this step splitting train and test into( 2 input variable of train and test also 2output varaibles of test and train)
```

```python
from sklearn.linear_model import LinearRegression#We founded a model is linear model so from using sklearn creating a model as 
#linear_model and importing LinearRegression
regressor=LinearRegression() #assigning LinearRegression() class as variable to regressor
regressor.fit(X_train,Y_train) #calling regressor.fit by the training set i/p & o/p as(X_train,Y_train)
#substituting a values in formula in this step,learning process is done here also calculating values for w and b
#once the above line is executed model is created it is called the LinearRegression() (reference ↓ line) and values are saved in regressor
```

```python
weight=regressor.coef_#after model is created we are cross checking with weight assigning regressor.coeffient = variable weight and displaying weight
weight
```

```python
bias=regressor.intercept_ #after model is created we are cross checking with bias assigning regressor.intercept = variable bias and displaying bias
bias
```

```python
Y_pred=regressor.predict(X_test)#this is a evaluation process.in regressor all will be saved and in this calling predict as a function in regressor and evaluation metric for Xtest set since( ytest is actual data)
```

```python
from sklearn.metrics import r2_score #sklearn is a library and metric is folder and importing r2score is a function
r_score=r2_score(Y_test,Y_pred)#in this step assigning r2score to rscore with (actual value ,predicted value)
r_score#displaying rscore variable and ouput is nearly to 1 is good and it is near to zero is bad,in this case it is a good model
```

```python
import pickle#creating a function with pickle to save the model
filename="Finalized_model_linear.sav" #mentioned with .sav filename and assigned to filename variable
```

```python
pickle.dump(regressor,open(filename,'wb'))#opening a file with dump in the regressor which all is saved in regressor and open with (filename ,wb) wb=write binary
#here opeining a file with filename usimng pickle.dump and it is saved as finalized model linear in the same folder
```

```python
loaded_model=pickle.load(open("Finalized_model_linear.sav",'rb'))#here loading a saved model using load and open ("filename with ext", read binary)
result=loaded_model.predict([[15]])#after loaded, assigning loaded model.predict to result for checking with input
result displaying result
```
